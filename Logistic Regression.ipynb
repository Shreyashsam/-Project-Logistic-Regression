```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Logistic Regression train.csv",delimiter=',',skipinitialspace=True)
df
```

```python
del df['Name']
```

```python
df[df.Sex=='female']=1
df[df.Sex=='male']=0
```

```python
df
```

```python
y_train=df['Survived']
```

```python
del df['Survived']
```

```python
x_train=df
x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
data_test=pd.read_csv("Logistic Regression Test.csv",delimiter=',')
data_test
```

```python
del data_test['Name']
data_test[data_test.Sex=='female']=1
data_test[data_test.Sex=='male']=0
```

```python
data_test
```

```python
y_predict=model.predict(data_test)
```

```python
y_predict=np.array(y_predict)
# np.savetxt("Logictic_Reg_prediction",y_predict,delimiter=',',fmt='%f')
y_predict
```
